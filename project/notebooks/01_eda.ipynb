# EDA контрольного датасета InterviewTrainer AI
Ноутбук исследует размеченный набор, используемый для сравнения baseline и production AI-пайплайна.

```python
import json
import statistics
from collections import Counter
from pathlib import Path

root = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
rows = json.loads((root / 'data' / 'experiment_dataset.json').read_text(encoding='utf-8'))
len(rows)
```

```python
categories = Counter(row['category'] for row in rows)
types = Counter(row['expected_type'] for row in rows)
lengths = [len(row['text']) for row in rows]
{'categories': dict(categories), 'types': dict(types), 'mean_length': statistics.mean(lengths)}
```

```python
quality = {
    'missing_values': sum(not row.get(key) for row in rows for key in ('id', 'text', 'expected_type', 'category')),
    'duplicate_ids': len(rows) - len({row['id'] for row in rows}),
    'duplicate_texts': len(rows) - len({row['text'] for row in rows}),
}
quality
```
