## 1. Импорт библиотеки NumPy

```python
import numpy as np
```

## 2. Примеры создания массивов

### 2.1. Из обычных списков

```python
# Создание массива из обычного списка
list_array = np.array([1, 2, 3, 4])
print("Массив из списка:", list_array)

# Создание массивов с помощью встроенных функций
zeros_array = np.zeros(5)
ones_array = np.ones((2, 3))
arange_array = np.arange(0, 10, 2)

print("zeros:", zeros_array)
print("ones:\n", ones_array)
print("arange:", arange_array)
```

```python
# из двумерного списка
b = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
print("Двумерный массив из двумерного списка: \n", b, "\nТип: ", type(b), "\nФорма: ", b.shape)
```

2.2 Атрибуты массива — shape и dtype

```python
arr = np.array([[1.0, 2.0], [3.0, 4.0]])
print("Массив:\n", arr)
print("Форма (shape):", arr.shape)
print("Тип данных (dtype):", arr.dtype)
```

## 3. Индексирование и срезы

```python
arr = np.array([[10, 20, 30], [40, 50, 60]])

# Индексирование
print("Элемент [0, 1]:", arr[0, 1])

# Срезы
print("Первая строка:", arr[0, :])
print("Первый столбец:", arr[:, 0])
print("Подмассив:", arr[0:2, 1:3])

# Булево индексирование
mask = arr > 30
print("Элементы > 30:", arr[mask])
```

**4. Векторные операции и базовые статистики**

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Поэлементные арифметические операции
print("Сложение:", a + b)
print("Умножение:", a * b)
print("Квадрат:", a ** 2)

# Статистические функции
arr = np.array([[1, 2, 3], [4, 5, 6]])

print("Среднее:", np.mean(arr))
print("Сумма:", np.sum(arr))
print("Минимум:", np.min(arr))
print("Максимум:", np.max(arr))
print("Стандартное отклонение:", np.std(arr))

# По осям
print("Среднее по строкам:", np.mean(arr, axis=1))
print("Сумма по столбцам:", np.sum(arr, axis=0))
```
